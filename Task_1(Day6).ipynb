```python
#3. Standardize payment_mode and calculate the percentage share of each mode per city
import pandas as pd

df['payment_mode'] = (
    df['payment_mode']
    .str.strip()
    .str.lower()
)
```

```python
counts = (
    df.groupby(['city', 'payment_mode'])
      .size()
      .reset_index(name='count')
)
```

```python
counts['percentage_share'] = (
    counts['count'] /
    counts.groupby('city')['count'].transform('sum')
) * 100
```

```python
#4.Detect customers whose customer_age is invalid but have order amounts above the dataset median
import pandas as pd
df=pd.read_csv(r"C:\Tekworks\Day 6\Dataset_1.csv")
df['order_amount'] = pd.to_numeric(df['order_amount'], errors='coerce')
df['order_amount']
```

```python
median_order_amount = df['order_amount'].median()
median_order_amount 
```

```python
invalid_age_condition = (
    df['customer_age'].isna() |
    (df['customer_age'] <= 0) |
    (df['customer_age'] > 120)
)
invalid_age_condition
```

```python
suspicious_customers = df[
    invalid_age_condition &
    (df['order_amount'] > median_order_amount)
]
print(suspicious_customers)
```

```python
#5.Find duplicate customer_ids where the same customer placed orders in multiple cities
duplicate_customers = (
    df.groupby('customer_id')['city']
      .nunique()
      .reset_index(name='city_count')
)

result = duplicate_customers[duplicate_customers['city_count'] > 1]

print(result)
```

```python
multi_city_orders = df[
    df['customer_id'].isin(result['customer_id'])
].sort_values('customer_id')

print(multi_city_orders)
```

```python
#7.After cleaning, compute the average order value per payment mode, excluding invalid amounts
import pandas as pd
df['order_amount'] = pd.to_numeric(df['order_amount'], errors='coerce')
# Keep only valid (positive) amounts
clean_df = df[df['order_amount'] > 0]
clean_df
```

```python
#11.Convert performance_score into numeric by mapping text values to numbers
import pandas as pd
# sample mapping
performance_map = {
    "Excellent": 5,
    "Very Good": 4,
    "Good": 3,
    "Average": 2,
    "Poor": 1
}
# standardize text first (recommended)
df['performance_score'] = (
    df['performance_score']
    .str.strip()
    .str.title()
)
# map text to numeric
df['performance_score_numeric'] = df['performance_score'].map(performance_map)

print(df[['performance_score', 'performance_score_numeric']])
```

```python
#12.Clean experience_years by converting text values to numeric and removing negatives.
import pandas as pd
import numpy as np
# convert text to numeric (invalid values → NaN)
df['experience_years'] = pd.to_numeric(df['experience_years'], errors='coerce')
# remove negative values
df.loc[df['experience_years'] < 0, 'experience_years'] = np.nan
```

```python
df['experience_years'].describe()
```

```python
#13.Convert is_active values (yes/no/True/False) into boolean format. 
df['is_active'] = df['is_active'].astype(str).str.strip().str.lower()
# map to boolean
active_map = {
    'yes': True,
    'true': True,
    '1': True,
    'no': False,
    'false': False,
    '0': False
}
df['is_active'] = df['is_active'].map(active_map)
df['is_active']
```

```python
#14.Identify and remove duplicate records based on user_id.
df[df.duplicated(subset='user_id', keep=False)]
# remove duplicates (keep first occurrence)
df = df.drop_duplicates(subset='user_id', keep='first')
df
```

```python
#15.Create a clean final DataFrame containing only valid, standardized records.
# remove rows with critical missing values
clean_df = df.dropna(subset=[
    'user_id',
    'experience_years',
    'performance_score_numeric',
    'is_active'
])
# reset index
clean_df = clean_df.reset_index(drop=True)
clean_df
```
